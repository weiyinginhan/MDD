```python
import subprocess

def start_autogenstudio_ui(port):
    try:
        # Construct the command to start AutoGenStudio UI on the specified port
        command = ["autogenstudio", "ui", "--port", str(port)]
        
        # Use subprocess to run the command
        process = subprocess.Popen(command)
        
        # Print the process ID for reference
        print(f"AutoGenStudio UI started on port {port} with PID: {process.pid}")
        
    except Exception as e:
        print(f"An error occurred: {e}")

# Start AutoGenStudio UI on port 8081
start_autogenstudio_ui(8081)
```

```python
start = datetime(2024, 4, 1)  
end = datetime(2024, 5, 31)  
start = "2024-04-01"
end = "2024-05-31"

# Fetch data for NVDA and TSLA
nvda = yf.download("NVDA", start=start, end=end)
tesla = yf.download("TSLA", start=start, end=end)

# Display the fetched data
print("NVDA Data:\n", nvda)
print("Tesla Data:\n", tesla)
```

```python
plt.figure(figsize=(14, 7))  
plt.plot(nvda['Close'], label='NVDA')  
plt.plot(tesla['Close'], label='TSLA')  
plt.title('NVDA vs TESLA Stock Price in 2023')  
plt.xlabel('Date')  
plt.ylabel('Stock Price in USD')  
plt.legend()  
plt.savefig('nvda_tesla.png')  
plt.show()  
```
